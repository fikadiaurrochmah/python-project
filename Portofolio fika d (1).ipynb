# Exploratory Data Analysis

```python
#calculate frequency of purchase for each customer, then check top 10 customers with most frequently purchase
top_buyer = store.groupby(by=['customer_id','customer_name'], as_index=False)['quantity'].count()
print('The TOP 10 customers with most frequently purchase:')
top_buyer.rename(columns={'quantity' : 'total_order'}, inplace=True)
top_buyer['rank']=top_buyer['total_order'].rank(ascending=False)
top_buyer.sort_values(by='total_order', ascending=False).head(10)
```

```python
top_buyer.info()
```

```python
#Do the high frequent customers are contributing more revenue
revenue = store.groupby(by=['customer_id','customer_name'], as_index=False)['sales','profit'].sum()
revenue['rank_sales']=revenue['sales'].rank(ascending=False)
revenue['rank_profit']=revenue['profit'].rank(ascending=False)
revenue.sort_values(by=['sales', 'profit'], ascending=False).head(10)
```

```python
result = top_buyer.merge(revenue)
print('Do the high frequent customers are contributing more revenue? Are they also profitable - what is the profit margin across the buckets?')
result.sort_values(by='rank').head(10)
```

```python
#Which customer segment is most profitable in each year?
profitable = store[['order_date','segment', 'profit']].copy()
profitable['Year']=profitable['order_date'].dt.year
profitable_segment=profitable.groupby(['Year','segment']).sum()
profitable_segment
```

```python
#How the customers are distributed across the countries

cust_in_countries = store.groupby(['country','segment'])['customer_id'].count()
cust_in_countries
```

```python
cust= store[['country','segment','customer_id','sales', 'profit']].copy()
xx=cust.groupby(['country','segment'])[['sales','profit']].sum()
xxx=xx.join(cust_in_countries)
xxx.rename(columns={'customer_id' : 'total_customer'}, inplace=True)
print('Distribution of customers:')
xxx
```

```python
top_country = xxx[(xxx['total_customer']>100) & (xxx['profit']>10000)]
top_country
```

```python
#which country has top sales
top_sales = store.groupby(['country'])[['sales']].sum().reset_index().sort_values('sales', ascending=False)
top_sales.head(10)
```

```python
# top 5 profit-making product types on a yearly basis
top_product = store[['order_date','product_id', 'product_name', 'category','sub_category','profit']].copy()
top_product['year'] = store['order_date'].dt.year
top_product_profit = top_product.groupby(['year','product_id'])['profit'].sum()
top_product_profit
```

```python
product = top_product.set_index('order_date')
product.drop('year',axis=1,inplace=True)
aa=product.groupby('product_id')['profit'].resample('Y').sum()
aa
```
